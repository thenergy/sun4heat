```python
import pandas as pd
import matplotlib.pyplot as plt

header = ['Calentador 1', 'Calentador 2', 'Calentador 3', 'Calentador 205', 'Calentador 206','Calentador 207']

mnths = {1:'Enero',
         2:'Febrero',
         3:'Marzo',
         4:'Abril',
         5:'Mayo',
         6:'Junio',
         7:'Julio',
         8:'Agosto',
         9:'Septiembre',
         10:'Octubre',
         11:'Noviembre',
         12:'Diciembre'}

#Estaciones desde día 1 hasta 365
est = ['Verano','Otoño','Invierno','Primavera']
estaciones = {
    'Verano':{'begin':355,'finish':78},
    'Otoño' :{'begin':79,'finish':171},
    'Invierno':{'begin':172,'finish':264},
    'Primavera':{'begin':265,'finish':354}
               }

path = '/home/diego/Documentos/sun4heat/'
base = pd.read_excel(path + 'cons_cal_quem.xlsx', names = header, header = None )

base.index = pd.date_range(start='2020-04-01 08:00', end='2022-04-26 06:00', freq='H')



```

```python
##Agrupación anual / mensual
base_monthly = base.groupby([base.index.year,base.index.month]).mean()



##Agrupación mensual / horaría (hora 1 del mes 1 se agrupa)
base_hourly = base.groupby([base.index.month,base.index.hour]).mean()



##Agrupación mensual / día (día 1 del mes 1 se agrupan)
base_day = base.groupby([base.index.month,base.index.day]).mean()
base_day = base_day.reset_index()
base_day['Meses'] = base_day.level_1.map(mnths)

#base_year = base.groupby([base.index.year]).sum()


#print(base_monthly)

#print(base_hourly)

print(base_day)

#print(base_year)

```
